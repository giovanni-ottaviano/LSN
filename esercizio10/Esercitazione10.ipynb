## Esercizio 10.1: <span style="color:brown"> Simulated Anneling </span>
Si vuole risolvere il problema del commesso viaggiatore (TSP) utilizzando un diverso tipo di algoritmo metaeuristico: il **Simulated Anneling** (SA). Questo metodo consiste nell'utilizzare la funzione costo del problema, cioè la lunghezza del percorso $L^{(1)}$, come l'energia di un sistema termodinamico (meccanica statistica) e di campionare la rispettiva distribuzione di Boltzmann $p = exp\,(-\beta L(x))$. In questo caso $\beta$ è l'inverso di una temperatura fittizia, il cui progressivo aumento permette di far arrivare il sistema al minimo globale della funzione costo. Questa diminuzione di temperatura deve essere abbastanza lenta da mantenere l'ergodicità del sistema. 

###  Città distribuite su una circonferenza
Per questa prima simulazione sono state scelte casualmente $N_{cities} = 32$ città su una circoferenza di raggio $r=1$.\
Si riportano i grafici per la lunghezza del percorso ad ogni step e un grafico del miglior percorso. Inoltre si confrontano i risultati con quelli ottenuti per il miglior percorso di ogni generazione dell'algoritmo genetico.

```python
#load data
L1_SA = np.loadtxt('lengths_circle_SA.dat', dtype=float, max_rows=150000)
L1_GA = np.loadtxt('fitness_circle.dat', dtype=float)
MC_steps_SA = np.array([i for i in range(len(L1_SA))])
MC_steps_GA = np.array([i for i in range(len(L1_GA))])

MC_steps = [MC_steps_SA, MC_steps_GA]
L1 = [L1_SA, L1_GA]
labels = ["Path length - Simulate Anneling", "Path length - Genetic Algorithm"]
x_labels = [r"MC steps", r"Generation"]

#make plot
fig, axx = plt.subplots(1, 2, figsize=(22,7))
for ax, steps, L, x_l, label in zip(axx, MC_steps, L1, x_labels, labels):
    ax.plot(steps, L)
    ax.ticklabel_format(axis='x', style='sci',scilimits=(0,0))
    ax.set_title(label)
    ax.set_xlabel(x_l)
    ax.set_ylabel(r'$L^{(1)}$')
    ax.grid(True)
```

```python
x, y = np.loadtxt('path_circle_SA.dat', dtype=float, unpack=True)
x1, y1 = np.loadtxt('path_circle.dat', dtype=float, unpack=True)

x = np.append(x, x[0]) #add fist city to complete path
y = np.append(y, y[0])
x1 = np.append(x1, x1[0]) #add fist city to complete path
y1 = np.append(y1, y1[0])

xx = [x, x1]
yy = [y, y1]
labels = ["Best path - Simulated Anneling", "Best path - Genetic Algorithm"]

#print length of the best path
print('Lunghezza del miglior percorso (SA):', L1_SA[-1])
print('Lunghezza del miglior percorso (GA):', L1_GA[-1])
print()

#make plots
fig, axx = plt.subplots(1, 2, figsize=(14,5))
for ax, X, Y, label in zip(axx, xx, yy, labels):
    fig.subplots_adjust(wspace=0.3)
    ax.plot(X, Y, marker='*')
    ax.set_title(label)
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')
    ax.grid(True)
```

## Esercizio 10.2: <span style="color:brown"> Parallelizzare l'Algoritmo Genetico usando MPI </span>
Si estende l'Algoritmo Genetico in modo da poterlo parallelizzare utilizzando la libreria *Message Passing Interface* (MPI).\
Il paradigma utilizzato è quello del *Single Program Multiple Data* (SPMD), che consiste nel far girare lo stesso programma su diversi nodi di calcolo, utilizzando diversi set di dati; in questo modo ciascun nodo esegue un GA indipendente, dato che i generetori di numeri casuali sono inizializzato con semi diversi.\
Ogni $N_{migr}$ steps dell'algoritmo (generazioni) i nodi comunicano tra loro per scambiarsi il miglior individuo; la coppia che invia e riceve è scelta casualmente.

Il setting della simulazione è uguale a quello utilizzato per la seconda parte dell'**Esercizio 09.1**, cioè sono state prese $N_{cities} = 32$ città all'interno di un quadrato di lato $l=2$ e tutte le probabilità di mutazione e crossover sono state scelte uguali. L'unico valore che è stato cambiato è la l'esponente della roulette truccata (usata per l'operatore di selezione), che è stato scelto come $p=3$.\
Per questa simulazione sono stati usati $4\,cores$ e si è scelto $N_{migr} = 50$.

```python
#load data - rank 0 best path at each iteration
L1_PGA = np.loadtxt('fitness_square_parallel.dat', dtype=float)
MC_steps_PGA = np.array([i for i in range(len(L1_PGA))])

MC_steps = [MC_steps_PGA, MC_steps_GA]
L1 = [L1_PGA, L1_GA]
labels = ["Path length - Parallel Genetic Algorithm", "Path length - Genetic Algorithm"]

#make plot
fig, axx = plt.subplots(1, 2, figsize=(22,7))
for ax, steps, L, x_l, label in zip(axx, MC_steps, L1, x_labels, labels):
    ax.plot(steps, L)
    ax.ticklabel_format(axis='x', style='sci',scilimits=(0,0))
    ax.set_title(label)
    ax.set_xlabel(r'Generation')
    ax.set_ylabel(r'$L^{(1)}$')
    ax.grid(True)
```

```python
#load data for parallel GA
x2, y2 = np.loadtxt('path_square_parallel.dat', dtype=float, unpack=True)

x2 = np.append(x2, x2[0]) #add fist city to complete path
y2 = np.append(y2, y2[0])

xx = [x2, x1]
yy = [y2, y1]
labels = ["Best path - Parallel Genetic Algorithm", "Best path - Genetic Algorithm"]

#print length of the best path
print('Lunghezza del miglior percorso (PGA):', round(L1_PGA[-1], 5))
print('Lunghezza del miglior percorso (GA): ', L1_GA[-1])
print()

#make plots
fig, axx = plt.subplots(1, 2, figsize=(14,5))
for ax, X, Y, label in zip(axx, xx, yy, labels):
    fig.subplots_adjust(wspace=0.3)
    ax.plot(X, Y, marker='*')
    ax.set_title(label, size=18)
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')
    ax.grid(True)
```

Si nota che l'utilizzo del calcolo parallelo e delle migrazioni permette di ottenere risultati migliori del singolo Algoritmo Genetico (e del Simulated Anneling); questo era un comportamento atteso, visto che le migrazioni tra un *continente* e l'altro permettono di attuare più mutazioni e crossover sugli individui migliori che dopo alcune migrazioni si andranno a riprodurre in tutti i continenti.
